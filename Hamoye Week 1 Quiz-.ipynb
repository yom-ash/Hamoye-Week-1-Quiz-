```python
fuel_type_group =fuel_data.groupby('fuel_type_code_pudl')
fuel_type_group= fuel_type_group.mean()
fuel_type_group
```

```python
fuel_cost_per_unit = fuel_type_group.sort_values(['fuel_cost_per_unit_burned'], ascending=False)
fuel_cost_per_unit
```

### Standard Deviation

```python
standard = fuel_data['fuel_mmbtu_per_unit'].std()
standard
```

### Percentile

```python
percentile = fuel_data['fuel_mmbtu_per_unit'].describe(percentiles = [0.75])
percentile
```

```python
#Percentage Change

#Group by report year and fuel type
year_cost = fuel_data.groupby(['report_year', 'fuel_type_code_pudl'] )
#Find mean of group
year_cost = year_cost.mean()
#Identify specific group i.e fuel cost per unit burned
yearscost = year_cost['fuel_cost_per_unit_burned']
```

```python
#Sort out specific years
year1994,year1998= yearscost.loc[[1994, 1998], 'coal']
#Calculate percentage change
percentchange = ((year1998-year1994)/year1994)*100
percentchange
```

```python
#Year had the most fuel quantity burned

#Group by year
group_year=fuel_data.groupby('report_year')
#Find mean of group
group_year = group_year.mean()
#Sort group in ascending order by fuel cost per unit
fuel_cost_perunit= group_year.sort_values(['fuel_cost_per_unit_delivered'], ascending = True)
fuel_cost_perunit
```
